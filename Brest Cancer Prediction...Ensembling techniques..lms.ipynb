### Compute and plot those feature which are related to each other?

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr =df.corr()
corr
```

```python
plt.figure(figsize=(30, 15))
sns.heatmap(corr[(corr >= 0.6) | (corr <= -0.7)], 
            cmap='YlGnBu', vmax=1.0, vmin=-1.0,
            annot=True, annot_kws={"size": 15})
plt.title('Correlation between features', fontsize=15)
plt.show()
```

### Replace M and B to 1 and 0 in diagnosis

```python
df.loc[df['diagnosis']=='M','diagnosis']=1
df.loc[df['diagnosis']=='B','diagnosis']=0
```

### changing datatype of diagnosis

```python
df['diagnosis']=df['diagnosis'].astype('float')
```

### independent and dependent variable

```python
x=df.drop(['diagnosis','Unnamed: 32'],axis=1)
y=df.diagnosis
```

### Build a ensemle learning model using bagging technique ?

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2, random_state=42)
from sklearn.ensemble import BaggingClassifier
from sklearn import tree
from sklearn import metrics
# build the model
meta_estimator = BaggingClassifier(tree.DecisionTreeClassifier(random_state=10))
# fit the model
meta_estimator.fit(x_train, y_train)
y_pred = meta_estimator.predict(x_test)
bagging_score= metrics.roc_auc_score(y_test, y_pred)
bagging_score
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
accuracy
```

### area under the roc curve and plot 

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.patches as mpatches
# set the figure size
plt.rcParams['figure.figsize']=(8,5)

fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# plot the ROC curve
plt.plot(fpr,tpr,label=' AUC Score = %0.3f'%(bagging_score))

# set limits for x and y axes
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])

# plot the straight line showing worst prediction for the model
plt.plot([0, 1], [0, 1],'r--')

# add the AUC score
plt.xlabel('False positive rate (1-Specificity)')
plt.ylabel('True positive rate (Sensitivity)')

plt.legend(loc='lower right')

plt.grid(True)
```

### Build a model using AdaBoost model

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
# build the model
adaboost = AdaBoostClassifier(random_state=10)
# fit the model
adaboost.fit(x_train, y_train)
# predict the values
y_pred_adaboost  = adaboost.predict(x_test)
adaboost_score = metrics.roc_auc_score(y_test, y_pred_adaboost)
adaboost_score
```

```python
from sklearn.metrics import accuracy_score
accuracy2=accuracy_score(y_test,y_pred_adaboost)
accuracy2
```
